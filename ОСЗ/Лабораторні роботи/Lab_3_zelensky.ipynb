<p>3. Імпортувати RGB-зображення Image з диску. Оцінити його розмір у пікселях <i>N x M</i> та об’єм пам’яті, який він займає на диску.</p>

<p>4. Імпортувати RGB-зображення Image у напівтоновому форматі. Оцінити його розмір у пікселях та об’єм пам’яті, який він займає на диску. Порівняти результати з попередніми результатами. Пояснити результати.</p>

```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import numpy as np

image_path = 'img/photo.png' 

image_rgb = cv.imread(image_path)

image_gray = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

if image_rgb is None or image_gray is None:
    print(f"Не вдалося знайти або відкрити зображення за шляхом: {image_path}")
else:
    print("(RGB)")
    height_N_rgb = image_rgb.shape[0]
    width_M_rgb = image_rgb.shape[1]
    channels_rgb = image_rgb.shape[2]
    print(f"Розмір у пікселях (N x M): {height_N_rgb} x {width_M_rgb}")
    print(f"Кількість каналів: {channels_rgb}")
    
    file_size_bytes_rgb = os.path.getsize(image_path)
    print(f"Об'єм на диску (оригінал): {file_size_bytes_rgb} байт\n")

    print("(Напівтонове)")
    height_N_gray = image_gray.shape[0]
    width_M_gray = image_gray.shape[1]
    print(f"Розмір у пікселях (N x M): {height_N_gray} x {width_M_gray}")
    print(f"Кількість каналів: 1")

    gray_image_path = 'img/gray_photo.png'
    cv.imwrite(gray_image_path, image_gray)
    file_size_bytes_gray = os.path.getsize(gray_image_path)
    print(f"Об'єм на диску (напівтоновий): {file_size_bytes_gray} байт")

    print("\n--- Порівняння ---\n")

    print(f"Розмір у пікселях (N x M):")
    print(f"  - RGB: {height_N_rgb} x {width_M_rgb}")
    print(f"  - Напівтонове: {height_N_gray} x {width_M_gray}")
    
    print(f"Об'єм на диску:")
    print(f"  - RGB (оригінал): {file_size_bytes_rgb} байт")
    print(f"  - Напівтонове (нове): {file_size_bytes_gray} байт")

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image_rgb, cv.COLOR_BGR2RGB))
    plt.title('Завдання 3: RGB')
    
    plt.subplot(1, 2, 2)
    plt.imshow(image_gray, cmap='gray')
    plt.title('Завдання 4: Напівтонове')
    plt.show()
```

<p>1. Оригінальний файл є дуже добре оптимізованим (лише 27KB).</p>

<p>5. Зменшити початкове RGB-зображення Image у n разів, візуалізувати його, перевіривши результат масштабування, та експортувати його на диск. Чи змінився об’єм пам’яті, який він займає на диску? Обгрунтувати отриманий результат.</p>

```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import math

image_path = 'img/photo.png' 
n = 2

image_rgb = cv.imread(image_path)

if image_rgb is None:
    print(f"Не вдалося знайти зображення за шляхом: {image_path}")
else:
    height_N, width_M, _ = image_rgb.shape
    file_size_bytes_original = os.path.getsize(image_path)

    print(f"Завдання 5")
    print(f"Оригінальний розмір (N x M): {height_N} x {width_M}")
    print(f"Оригінальний об'єм на диску: {file_size_bytes_original} байт\n")

    new_width = math.floor(width_M / n)
    new_height = math.floor(height_N / n)
    new_dimensions = (new_width, new_height)

    image_resized = cv.resize(image_rgb, new_dimensions, interpolation=cv.INTER_AREA)
    
    print(f"Зображення зменшено у {n} рази.")
    print(f"Новий розмір (N x M): {new_height} x {new_width}\n")

    resized_image_path = 'img/photo_res.png'
    cv.imwrite(resized_image_path, image_resized, [cv.IMWRITE_PNG_COMPRESSION, 9])
    print(f"Зменшене зображення збережено як: {resized_image_path}")

    file_size_bytes_resized = os.path.getsize(resized_image_path)
    print(f"Новий об'єм на диску: {file_size_bytes_resized} байт\n")

    if file_size_bytes_resized < file_size_bytes_original:
        print(f"Об'єм пам'яті на диску зменшився.")
        print(f"Зменшення склало: {file_size_bytes_original - file_size_bytes_resized} байт.")
    else:
        print("Об'єм пам'яті не зменшився, що незвично для зменшення.")

        print(f"Всупереч очікуванням, новий файл ({file_size_bytes_resized} байт) виявився **більшим** за оригінал ({file_size_bytes_original} байт), \n"
          f"навіть попри те, що він має значно менше пікселів ({new_height * new_width} проти {height_N * width_M}).\n\n"
          f"Це пояснюється тим, що на розмір файлу впливає не лише кількість пікселів, \n"
          f"а й **алгоритм та ефективність стиснення**:\n\n"
          f"1. Оригінальний файл ('{image_path}') вочевидь був збережений з **високою оптимізацією** \n"
          f"   (наприклад, зі зменшенням палітри кольорів), що дозволило досягти такого малого розміру.\n\n"
          f"2. Функція `cv.imwrite()` використала **стандартний (менш ефективний) алгоритм стиснення PNG**.\n\n"
          f"**Висновок:** Файл із меншою кількістю пікселів, але **без сильної оптимізації**, \n"
          f"зайняв на диску більше місця, ніж високооптимізований оригінал.")

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image_rgb, cv.COLOR_BGR2RGB))
    plt.title(f'Оригінал: {width_M} x {height_N}')
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_resized, cv.COLOR_BGR2RGB))
    plt.title(f'Зменшено (n={n}): {new_width} x {new_height}')
    
    plt.show()
```

<p>Всупереч очікуванням, новий файл 44224 байт) виявився більшим за оригінал 27881 байт, навіть попри те, що він має значно менше пікселів 22 500 проти 90 000. Це пояснюється тим, що на розмір файлу впливає не лише кількість пікселів, а й алгоритм та ефективність стиснення:</p>

<p>1. Оригінальний файл вочевидь був збережений з високою оптимізацією (наприклад, зі зменшенням палітри кольорів), що дозволило досягти такого малого розміру.</p>

<p>2. Функція cv.imwrite() використала стандартний (менш ефективний) алгоритм стиснення PNG.</p>

<p>Висновок: Файл із меншою кількістю пікселів, але без сильної оптимізації, зайняв на диску більше місця, ніж високооптимізований оригінал.</p><br>

<p>6. Написати функцію, яка буде виконувати пропорційне зменшення початкового зображення розміром N x M пікселей відповідно до вимоги, щоби зменешене зображення мало ширину K пікселей.</p>

```python
import cv2 as cv
import math
import matplotlib.pyplot as plt
import os

def resize_proportional(image, k_width):
    (height_N, width_M) = image.shape[:2]
    ratio = k_width / float(width_M)
    new_height = int(height_N * ratio)
    new_dimensions = (k_width, new_height)
    resized_image = cv.resize(image, new_dimensions, interpolation=cv.INTER_AREA)
    
    return resized_image

image_path = 'img/photo.png'
K = 200

image_rgb = cv.imread(image_path)

if image_rgb is None:
    print(f"Не вдалося знайти зображення за шляхом: {image_path}")
else:
    image_resized_k = resize_proportional(image_rgb, K)
    
    print(f"Результати Завдання 6")
    print(f"Оригінальний розмір (M x N): {image_rgb.shape[1]} x {image_rgb.shape[0]}")
    print(f"Новий розмір (K x ...):   {image_resized_k.shape[1]} x {image_resized_k.shape[0]}")

    plt.imshow(cv.cvtColor(image_resized_k, cv.COLOR_BGR2RGB))
    plt.title(f'K = {K}px')
    plt.show()
```

<p>7. Зробити фото розвороту власної залікової книжки з перспективним викривленням на фоні письмового столу. Вирізати зону із заліковкою та виконати афінні та перспективні перетворення за трьома точками точками і візуалізовані. В кожному випадку результати мають бути записані на диск.</p>

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

zalikova_path = 'img/zalikova.jpg' 
image = cv.imread(zalikova_path)

if image is None:
    print(f"Помилка: Не вдалося знайти фото за шляхом: {zalikova_path}")
else:
    print("Афінне перетворення")

    pts_input_affine = np.float32([
        [415, 340],
        [573, 342],
        [410, 540]
    ])

    pts_output_affine = np.float32([
        [220, 290],
        [470, 290],
        [200, 620]
    ])

    M_affine = cv.getAffineTransform(pts_input_affine, pts_output_affine)

    rows, cols, _ = image.shape
    image_affine = cv.warpAffine(image, M_affine, (cols, rows))

    cv.imwrite('img/zalikova_affine.png', image_affine)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title("Оригінал")
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_affine, cv.COLOR_BGR2RGB))
    plt.title("Афінне 3 точки")
    plt.show()
    
    print("Файл 'zalikova_affine.png' збережено.")
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

zalikova_path = 'img/zalikova.jpg' 
image = cv.imread(zalikova_path)

if image is not None:
    print("\nПерспективне перетворення")
    
    pts_input_perspective = np.float32([
        [220, 290],
        [470, 290],
        [200, 620],
        [465, 625]
    ])

    output_width = 400
    output_height = 500

    pts_output_perspective = np.float32([
        [0, 0],             
        [output_width, 0],     
        [0, output_height],  
        [output_width, output_height]
    ])

    M_perspective = cv.getPerspectiveTransform(pts_input_perspective, pts_output_perspective)

    image_perspective = cv.warpPerspective(image, M_perspective, (output_width, output_height))

    cv.imwrite('img/zalikova_perspective.png', image_perspective)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title("Оригінал")
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_perspective, cv.COLOR_BGR2RGB))
    plt.title("Перспективне")
    plt.show()
    
    print("Файл 'zalikova_perspective.png' збережено.")
```

<h2 style="font-weight:bold; font-size:22px;">Контрольні запитання</h2>

<ol>
    <li>У чому полягає сенс афінних перетворень?</li><br>
    <p>Сенс афінних перетворень полягає у збереженні паралельності. Це означає, що будь-які лінії, які були паралельними на початковому зображенні, залишаться паралельними і після трансформації. Ці перетворення є комбінацією лінійних операцій (масштабування, поворот, зсув) та трансляції (переміщення). Хоча вони зберігають прямі лінії, кути та відстані між точками можуть змінюватися.</p>
    <li>Чи є оперції трансляції, масштабування і повороту афінними? Відповідь обгрунтувати.</li><br>
    <p>Так, ці операції є окремими випадками афінних перетворень.</p>
    <p>Кожна з них підпорядковується головному правилу афінних перетворень — збереженню паралельності:</p>
    <ul>
    <li>Трансляція (зсув): Лише переміщує зображення; всі лінії зберігають свій кут нахилу та паралельність.</li>
    <li>Масштабування (зміна розміру): Збільшує або зменшує зображення, але паралельні лінії залишаються паралельними.</li>
    <li>Поворот: Обертає всі лінії на однаковий кут навколо центру, тому лінії, що були паралельними, залишаються паралельними і після повороту.</li>
    </ul>
    <li>Чи є перспективні перетворення частинним випадком афінних перетворень? Відповідь обгрунтувати.</li><br>
    <p>Ні, це не так. Насправді, афінне перетворення є більш обмеженим, частковим випадком перспективного.</p>
<ul>
    <li>Афінні перетворення (на основі 3 точок) завжди зберігають паралельність ліній.</li>
    <li>Перспективні перетворення (на основі 4 точок) не зберігають паралельність. Вони можуть змусити паралельні лінії "сходитися" в одній точці, імітуючи 3D-перспективу. Це дозволяє перетворити трапецію на прямокутник, що неможливо зробити лише афінним перетворенням.</li>
</ul>
</ol>
