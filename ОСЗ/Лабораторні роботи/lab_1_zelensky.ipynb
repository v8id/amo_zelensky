```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
print(f'version cv2: {cv.__version__}')

image = cv.imread('scr/photo.png')

print(f'Shape/Dimensions: {image.shape}')
print(f'Number of pixels: {image.size}')

gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

plt.figure(figsize=(12,6))

image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('Кольорове зображення')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title('Напівтонове зображення')
plt.axis('off')

plt.show()

cv.imwrite('scr/1234.jpg', gray)
```

<h2 style="font-weight:bold; font-size:22px;">Контрольні запитання</h2>
<ol>
    <li>Що таке pip і conda та чим вони відрізняються між собою?</li><br>
    <p>pip (Pip Installs Packages) — це офіційний менеджер пакетів для Python. Він встановлює пакети (бібліотеки) з репозиторію PyPI (Python Package Index). pip керує лише пакетами Python і для керування віртуальними середовищами потребує додаткових інструментів, як-от venv або virtualenv.</p>
    <p>conda — це одночасно менеджер пакетів та менеджер віртуальних середовищ. Він є частиною дистрибутиву Anaconda/Miniconda і встановлює пакети з репозиторіїв Anaconda. Головна відмінність conda в тому, що він може керувати пакетами для будь-яких мов (не лише Python, але й R, C++ тощо) та їхніми складними залежностями. conda також самостійно створює та керує ізольованими середовищами (conda create, conda activate).</p>
    <li>Як буде виглядати виконання пунктів 1-3 індивідуального завдання за допомогою Virtualenv?</li><br>
    <p>Спочатку створюємо віртуальне середовище командою: python -m venv myenv</p>
    <p>Потім активуємо його: myenv/Scripts/activate</p>
    <p>Встановити пакети: pip install numpy matplotlib opencv-python jupyter</p>

    <li>Як виконати візуалізацію зображення (див. п.3) засобами mathplotlib?</li><br>
    <p>Для коректної візуалізації зображення, завантаженого через OpenCV, за допомогою matplotlib, необхідно конвертувати його колірний простір. OpenCV (cv.imread) завантажує зображення у форматі BGR (Синій-Зелений-Червоний), тоді як matplotlib (plt.imshow) очікує формат RGB (Червоний-Зелений-Синій).</p>
    <pre><code>
import cv2 as cv
import matplotlib.pyplot as plt

# 1. Завантажити зображення (буде в BGR)
image_bgr = cv.imread('шлях/до/вашого/зображення.jpg')

# 2. Конвертувати з BGR в RGB
image_rgb = cv.cvtColor(image_bgr, cv.COLOR_BGR2RGB)

# 3. Відобразити за допомогою matplotlib
plt.imshow(image_rgb)
plt.title('Кольорове зображення')
plt.axis('off') # Вимкнути осі
plt.show()
</code></pre>
