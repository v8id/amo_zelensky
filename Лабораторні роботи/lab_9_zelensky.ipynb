```python
#Реалізування наївного пошуку шаблонів (The Naive String Matching Algorithm) для довільних рядків на python:
def naive_string_match(text, pattern):
    n = len(text)
    m = len(pattern)
    
    # Проходимо по всіх можливих позиціях тексту для пошуку шаблону
    for i in range(n - m + 1):  # Оскільки шаблон має довжину m, перевіряємо до n - m + 1
        # Перевіряємо, чи збігається шаблон з підрядком у поточній позиції
        if text[i:i + m] == pattern:
            print(f"Збіг знайдено на позиції {i}")
            
# Приклад використання
text = "ABABDABACDABABCABAB"
pattern = "ABAB"
naive_string_match(text, pattern)
```

```python
#Реалізування Z-функції для довільних рядків на python:
def z_function(s):
    n = len(s)
    z = [0] * n  # Массив Z-функції
    l, r, k = 0, 0, 0
    
    for i in range(1, n):
        if i <= r:
            z[i] = min(r - i + 1, z[i - l])  # Якщо i знаходиться в межах [l, r], використовуємо значення з попередніх обчислень
        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1  # Збільшуємо значення Z-функції, поки є збіги
        if i + z[i] - 1 > r:
            l, r = i, i + z[i] - 1  # Оновлюємо межі [l, r]
    
    return z

# Приклад використання:
s = "abacabadabacaba"
z_values = z_function(s)
print("Z-функція для рядка:", z_values)
```

```python
#Реалізація наївного пошуку підрядка:
def naive_search(text, pattern):
    n = len(text)
    m = len(pattern)
    
    # Перевірка кожної позиції в тексті для можливості знайти шаблон
    for i in range(n - m + 1):
        j = 0
        while j < m and text[i + j] == pattern[j]:
            j += 1
        if j == m:  # Якщо всі символи підрядка співпали
            return i  # Повертаємо індекс початку знайденого підрядка
    
    return -1  # Якщо підрядок не знайдений

# Приклад використання:
text = "abacabadabacaba"
pattern = "abad"
result = naive_search(text, pattern)
if result != -1:
    print(f"Підрядок знайдений на позиції: {result}")
else:
    print("Підрядок не знайдений")
```

```python
#Реалізація cтиснення рядка:
def compress_string(s):
    # Перевіряємо, чи рядок порожній
    if not s:
        return ""

    # Створюємо список для збереження результату
    result = []
    count = 1  # Лічильник кількості повторів символу

    # Проходимо по рядку
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:  # Якщо поточний символ дорівнює попередньому
            count += 1
        else:
            result.append(s[i - 1])  # Додаємо попередній символ
            if count > 1:  # Якщо символ повторюється більше одного разу
                result.append(str(count))  # Додаємо кількість повторів
            count = 1  # Скидаємо лічильник для наступного символу

    # Не забуваємо додати останній символ та його кількість
    result.append(s[-1])
    if count > 1:
        result.append(str(count))

    return ''.join(result)

# Приклад використання
input_string = "aaaaaaaaaaaaaaaaabbbbbbbbbccccdaaaaaaaaaaaaaaaa"
compressed_string = compress_string(input_string)
print(f"Оригінальний рядок: {input_string}")
print(f"Стиснутий рядок: {compressed_string}")
```

#### **2)** Пояснюємо асимптотику наївного алгоритму пошуку підрядка.
Наївний алгоритм пошуку підрядка має просту концепцію: ми шукаємо всі можливі позиції, на яких підрядок може починатися в рядку, і перевіряємо, чи збігається підрядок з частиною рядка на цій позиції.

**Опис алгоритму:**
- Ми починаємо з першої позиції в основному рядку (довжина $n$) і перевіряємо, чи збігається підрядок довжиною $m$ з частиною основного рядка.
- Якщо збіг є, то ми знаходимо підрядок. Якщо ні, переходимо до наступної позиції і повторюємо порівняння.
- Повторюємо цей процес для кожної з $n - m + 1$ можливих позицій.

**Оцінка складності:**
1. Часова складність:
- В найгіршому випадку ми повинні порівнювати всі символи підрядка з відповідними символами в основному рядку для кожної з позицій.
- Для кожної з $n - m + 1$ можливих позицій потрібно зробити $m$ порівнянь (де $n$ — довжина основного рядка, а $m$ — довжина підрядка).

Тому загальна кількість операцій у найгіршому випадку:$(n−m+1)×m=O(n×m)$

У найгіршому випадку (коли $m$ наближається до $n$) складність буде $O(n^2)$.

2. Просторова складність:
Алгоритм не вимагає додаткової пам'яті для зберігання даних, окрім введених рядків. Тому просторова складність: $O(1)$.

#### **1)** Що таке «префікс-функція» у контексті алгоритмів на рядках, і як вона відрізняється від Z-функції?
Префікс-функція обчислює для кожної позиції рядка довжину найбільшого власного префікса, який є одночасно суфіксом підрядка до цієї позиції (використовується в алгоритмі Кнута-Морріса-Пратта для пошуку підрядків).

Z-функція визначає для кожної позиції рядка довжину найдовшого префікса, що співпадає з підрядком, який починається з цієї позиції (використовується для пошуку входжень шаблону в рядок та стиснення).

Основні відмінності:
- Префікс-функція обчислює співпадіння префіксів і суфіксів до певної позиції, Z-функція — співпадіння з підрядками, що починаються з цієї позиції.
- Префікс-функція застосовується в алгоритмі Кнута-Морріса-Пратта, Z-функція — в пошуку шаблонів і стисненні рядків.

#### **2)** Що таке Z-функція у вищому контексті алгоритмів на рядках, і яка її роль у вирішенні задач?

Роль:

- Інші задачі: Застосовується у вирішенні задач на рядках, наприклад, у побудові префіксних дерев.

#### **3)** Які існують підходи до вирішення задачі «найдовший спільний підрядок» для двох рядків?
Існують три основні підходи для задачі «найдовший спільний підрядок»:

1. Динамічне програмування: Створюється матриця, де кожен елемент $DP[i][j]$ містить довжину найдовшого спільного підрядка для префіксів двох рядків. Часова складність: $O(n \times m)$.

2. Суфіксні масиви і LCP: Будуємо суфіксні масиви та масиви довгих спільних префіксів для рядків. Часова складність: $O(n \log n)$.

3. Бінарний пошук з хешуванням: Застосовуємо бінарний пошук для знаходження максимальної довжини спільного підрядка через хешування. Часова складність: $O(n \log n)$.

Метод вибирається в залежності від розміру рядків і вимог до ефективності.
