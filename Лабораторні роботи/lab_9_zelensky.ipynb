### **1.** –°—Ç–≤–æ—Ä—é—î–º–æ Notebook-–¥–æ–∫—É–º–µ–Ω—Ç —ñ —Ä–µ–∞–ª—ñ–∑–æ–≤—É—î–º–æ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ñ –ø—Ä–∏–∫–ª–∞–¥–∏, —â–æ —Ä–æ–∑–≥–ª—è–¥–∞—é—Ç—å—Å—è —É —Ü—ñ–π —Ä–æ–±–æ—Ç—ñ, —Ç–∞ –≤–∏–∫–æ–Ω—É—î–º–æ –∑–∞–≤–¥–∞–Ω–Ω—è, —â–æ –Ω–∞–¥–∞–Ω–æ –Ω–∞ —Å–∞–º–æ—Å—Ç—ñ–π–Ω—É —Ä–æ–±–æ—Ç—É.
### <center> –ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ—ó —Ä–æ–±–æ—Ç–∏ </center>

#### **1)** –ü–æ–≤—Ç–æ—Ä—é—î–º–æ –∫–æ–¥, –Ω–∞–≤–µ–¥–µ–Ω–∏–π –≤–∏—â–µ –¥–ª—è –¥–æ–≤—ñ–ª—å–Ω–∏—Ö —Ä—è–¥–∫—ñ–≤;

```python
#–†–µ–∞–ª—ñ–∑—É–≤–∞–Ω–Ω—è –Ω–∞—ó–≤–Ω–æ–≥–æ –ø–æ—à—É–∫—É —à–∞–±–ª–æ–Ω—ñ–≤ (The Naive String Matching Algorithm) –¥–ª—è –¥–æ–≤—ñ–ª—å–Ω–∏—Ö —Ä—è–¥–∫—ñ–≤ –Ω–∞ python:
def naive_string_match(text, pattern):
    n = len(text)
    m = len(pattern)
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º–æ –ø–æ –≤—Å—ñ—Ö –º–æ–∂–ª–∏–≤–∏—Ö –ø–æ–∑–∏—Ü—ñ—è—Ö —Ç–µ–∫—Å—Ç—É –¥–ª—è –ø–æ—à—É–∫—É —à–∞–±–ª–æ–Ω—É
    for i in range(n - m + 1):  # –û—Å–∫—ñ–ª—å–∫–∏ —à–∞–±–ª–æ–Ω –º–∞—î –¥–æ–≤–∂–∏–Ω—É m, –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –¥–æ n - m + 1
        # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –∑–±—ñ–≥–∞—î—Ç—å—Å—è —à–∞–±–ª–æ–Ω –∑ –ø—ñ–¥—Ä—è–¥–∫–æ–º —É –ø–æ—Ç–æ—á–Ω—ñ–π –ø–æ–∑–∏—Ü—ñ—ó
        if text[i:i + m] == pattern:
            print(f"–ó–±—ñ–≥ –∑–Ω–∞–π–¥–µ–Ω–æ –Ω–∞ –ø–æ–∑–∏—Ü—ñ—ó {i}")
            
# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
text = "ABABDABACDABABCABAB"
pattern = "ABAB"
naive_string_match(text, pattern)
```

```python
#–†–µ–∞–ª—ñ–∑—É–≤–∞–Ω–Ω—è Z-—Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è –¥–æ–≤—ñ–ª—å–Ω–∏—Ö —Ä—è–¥–∫—ñ–≤ –Ω–∞ python:
def z_function(s):
    n = len(s)
    z = [0] * n  # –ú–∞—Å—Å–∏–≤ Z-—Ñ—É–Ω–∫—Ü—ñ—ó
    l, r, k = 0, 0, 0
    
    for i in range(1, n):
        if i <= r:
            z[i] = min(r - i + 1, z[i - l])  # –Ø–∫—â–æ i –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤ –º–µ–∂–∞—Ö [l, r], –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ—Ö –æ–±—á–∏—Å–ª–µ–Ω—å
        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1  # –ó–±—ñ–ª—å—à—É—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è Z-—Ñ—É–Ω–∫—Ü—ñ—ó, –ø–æ–∫–∏ —î –∑–±—ñ–≥–∏
        if i + z[i] - 1 > r:
            l, r = i, i + z[i] - 1  # –û–Ω–æ–≤–ª—é—î–º–æ –º–µ–∂—ñ [l, r]
    
    return z

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è:
s = "abacabadabacaba"
z_values = z_function(s)
print("Z-—Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è —Ä—è–¥–∫–∞:", z_values)
```

```python
#–†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –Ω–∞—ó–≤–Ω–æ–≥–æ –ø–æ—à—É–∫—É –ø—ñ–¥—Ä—è–¥–∫–∞:
def naive_search(text, pattern):
    n = len(text)
    m = len(pattern)
    
    # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –∫–æ–∂–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó –≤ —Ç–µ–∫—Å—Ç—ñ –¥–ª—è –º–æ–∂–ª–∏–≤–æ—Å—Ç—ñ –∑–Ω–∞–π—Ç–∏ —à–∞–±–ª–æ–Ω
    for i in range(n - m + 1):
        j = 0
        while j < m and text[i + j] == pattern[j]:
            j += 1
        if j == m:  # –Ø–∫—â–æ –≤—Å—ñ —Å–∏–º–≤–æ–ª–∏ –ø—ñ–¥—Ä—è–¥–∫–∞ —Å–ø—ñ–≤–ø–∞–ª–∏
            return i  # –ü–æ–≤–µ—Ä—Ç–∞—î–º–æ —ñ–Ω–¥–µ–∫—Å –ø–æ—á–∞—Ç–∫—É –∑–Ω–∞–π–¥–µ–Ω–æ–≥–æ –ø—ñ–¥—Ä—è–¥–∫–∞
    
    return -1  # –Ø–∫—â–æ –ø—ñ–¥—Ä—è–¥–æ–∫ –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–∏–π

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è:
text = "abacabadabacaba"
pattern = "abad"
result = naive_search(text, pattern)
if result != -1:
    print(f"–ü—ñ–¥—Ä—è–¥–æ–∫ –∑–Ω–∞–π–¥–µ–Ω–∏–π –Ω–∞ –ø–æ–∑–∏—Ü—ñ—ó: {result}")
else:
    print("–ü—ñ–¥—Ä—è–¥–æ–∫ –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–∏–π")
```

```python
#–†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è c—Ç–∏—Å–Ω–µ–Ω–Ω—è —Ä—è–¥–∫–∞:
def compress_string(s):
    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ —Ä—è–¥–æ–∫ –ø–æ—Ä–æ–∂–Ω—ñ–π
    if not s:
        return ""

    # –°—Ç–≤–æ—Ä—é—î–º–æ —Å–ø–∏—Å–æ–∫ –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
    result = []
    count = 1  # –õ—ñ—á–∏–ª—å–Ω–∏–∫ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –ø–æ–≤—Ç–æ—Ä—ñ–≤ —Å–∏–º–≤–æ–ª—É

    # –ü—Ä–æ—Ö–æ–¥–∏–º–æ –ø–æ —Ä—è–¥–∫—É
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:  # –Ø–∫—â–æ –ø–æ—Ç–æ—á–Ω–∏–π —Å–∏–º–≤–æ–ª –¥–æ—Ä—ñ–≤–Ω—é—î –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ–º—É
            count += 1
        else:
            result.append(s[i - 1])  # –î–æ–¥–∞—î–º–æ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–π —Å–∏–º–≤–æ–ª
            if count > 1:  # –Ø–∫—â–æ —Å–∏–º–≤–æ–ª –ø–æ–≤—Ç–æ—Ä—é—î—Ç—å—Å—è –±—ñ–ª—å—à–µ –æ–¥–Ω–æ–≥–æ —Ä–∞–∑—É
                result.append(str(count))  # –î–æ–¥–∞—î–º–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–≤—Ç–æ—Ä—ñ–≤
            count = 1  # –°–∫–∏–¥–∞—î–º–æ –ª—ñ—á–∏–ª—å–Ω–∏–∫ –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ —Å–∏–º–≤–æ–ª—É

    # –ù–µ –∑–∞–±—É–≤–∞—î–º–æ –¥–æ–¥–∞—Ç–∏ –æ—Å—Ç–∞–Ω–Ω—ñ–π —Å–∏–º–≤–æ–ª —Ç–∞ –π–æ–≥–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å
    result.append(s[-1])
    if count > 1:
        result.append(str(count))

    return ''.join(result)

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
input_string = "aaaaaaaaaaaaaaaaabbbbbbbbbccccdaaaaaaaaaaaaaaaa"
compressed_string = compress_string(input_string)
print(f"–û—Ä–∏–≥—ñ–Ω–∞–ª—å–Ω–∏–π —Ä—è–¥–æ–∫: {input_string}")
print(f"–°—Ç–∏—Å–Ω—É—Ç–∏–π —Ä—è–¥–æ–∫: {compressed_string}")
```

#### **2)** –ü–æ—è—Å–Ω—é—î–º–æ –∞—Å–∏–º–ø—Ç–æ—Ç–∏–∫—É –Ω–∞—ó–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –ø–æ—à—É–∫—É –ø—ñ–¥—Ä—è–¥–∫–∞.
–ù–∞—ó–≤–Ω–∏–π –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ—à—É–∫—É –ø—ñ–¥—Ä—è–¥–∫–∞ –º–∞—î –ø—Ä–æ—Å—Ç—É –∫–æ–Ω—Ü–µ–ø—Ü—ñ—é: –º–∏ —à—É–∫–∞—î–º–æ –≤—Å—ñ –º–æ–∂–ª–∏–≤—ñ –ø–æ–∑–∏—Ü—ñ—ó, –Ω–∞ —è–∫–∏—Ö –ø—ñ–¥—Ä—è–¥–æ–∫ –º–æ–∂–µ –ø–æ—á–∏–Ω–∞—Ç–∏—Å—è –≤ —Ä—è–¥–∫—É, —ñ –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –∑–±—ñ–≥–∞—î—Ç—å—Å—è –ø—ñ–¥—Ä—è–¥–æ–∫ –∑ —á–∞—Å—Ç–∏–Ω–æ—é —Ä—è–¥–∫–∞ –Ω–∞ —Ü—ñ–π –ø–æ–∑–∏—Ü—ñ—ó.

**–û–ø–∏—Å –∞–ª–≥–æ—Ä–∏—Ç–º—É:**
- –ú–∏ –ø–æ—á–∏–Ω–∞—î–º–æ –∑ –ø–µ—Ä—à–æ—ó –ø–æ–∑–∏—Ü—ñ—ó –≤ –æ—Å–Ω–æ–≤–Ω–æ–º—É —Ä—è–¥–∫—É (–¥–æ–≤–∂–∏–Ω–∞ $n$) —ñ –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –∑–±—ñ–≥–∞—î—Ç—å—Å—è –ø—ñ–¥—Ä—è–¥–æ–∫ –¥–æ–≤–∂–∏–Ω–æ—é $m$ –∑ —á–∞—Å—Ç–∏–Ω–æ—é –æ—Å–Ω–æ–≤–Ω–æ–≥–æ —Ä—è–¥–∫–∞.
- –Ø–∫—â–æ –∑–±—ñ–≥ —î, —Ç–æ –º–∏ –∑–Ω–∞—Ö–æ–¥–∏–º–æ –ø—ñ–¥—Ä—è–¥–æ–∫. –Ø–∫—â–æ –Ω—ñ, –ø–µ—Ä–µ—Ö–æ–¥–∏–º–æ –¥–æ –Ω–∞—Å—Ç—É–ø–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó —ñ –ø–æ–≤—Ç–æ—Ä—é—î–º–æ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è.
- –ü–æ–≤—Ç–æ—Ä—é—î–º–æ —Ü–µ–π –ø—Ä–æ—Ü–µ—Å –¥–ª—è –∫–æ–∂–Ω–æ—ó –∑ $n - m + 1$ –º–æ–∂–ª–∏–≤–∏—Ö –ø–æ–∑–∏—Ü—ñ–π.

**–û—Ü—ñ–Ω–∫–∞ —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ:**
1. –ß–∞—Å–æ–≤–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å:
- –í –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É –º–∏ –ø–æ–≤–∏–Ω–Ω—ñ –ø–æ—Ä—ñ–≤–Ω—é–≤–∞—Ç–∏ –≤—Å—ñ —Å–∏–º–≤–æ–ª–∏ –ø—ñ–¥—Ä—è–¥–∫–∞ –∑ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–º–∏ —Å–∏–º–≤–æ–ª–∞–º–∏ –≤ –æ—Å–Ω–æ–≤–Ω–æ–º—É —Ä—è–¥–∫—É –¥–ª—è –∫–æ–∂–Ω–æ—ó –∑ –ø–æ–∑–∏—Ü—ñ–π.
- –î–ª—è –∫–æ–∂–Ω–æ—ó –∑ $n - m + 1$ –º–æ–∂–ª–∏–≤–∏—Ö –ø–æ–∑–∏—Ü—ñ–π –ø–æ—Ç—Ä—ñ–±–Ω–æ –∑—Ä–æ–±–∏—Ç–∏ $m$ –ø–æ—Ä—ñ–≤–Ω—è–Ω—å (–¥–µ $n$ ‚Äî –¥–æ–≤–∂–∏–Ω–∞ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ —Ä—è–¥–∫–∞, –∞ $m$ ‚Äî –¥–æ–≤–∂–∏–Ω–∞ –ø—ñ–¥—Ä—è–¥–∫–∞).

–¢–æ–º—É –∑–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü—ñ–π —É –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É:$(n‚àím+1)√óm=O(n√óm)$

–£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É (–∫–æ–ª–∏ $m$ –Ω–∞–±–ª–∏–∂–∞—î—Ç—å—Å—è –¥–æ $n$) —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –±—É–¥–µ $O(n^2)$.

2. –ü—Ä–æ—Å—Ç–æ—Ä–æ–≤–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å:
–ê–ª–≥–æ—Ä–∏—Ç–º –Ω–µ –≤–∏–º–∞–≥–∞—î –¥–æ–¥–∞—Ç–∫–æ–≤–æ—ó –ø–∞–º'—è—Ç—ñ –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö, –æ–∫—Ä—ñ–º –≤–≤–µ–¥–µ–Ω–∏—Ö —Ä—è–¥–∫—ñ–≤. –¢–æ–º—É –ø—Ä–æ—Å—Ç–æ—Ä–æ–≤–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å: $O(1)$.

#### **3)** –û–±—á–∏—Å–ª—é—î–º–æ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É —Å—Ç–∏—Å–Ω–µ–Ω–Ω—è —Ä—è–¥–∫–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é ùëß-—Ñ—É–Ω–∫—Ü—ñ—ó.

**–ö—Ä–æ–∫ 1:** –û–±—á–∏—Å–ª–µ–Ω–Ω—è Z-—Ñ—É–Ω–∫—Ü—ñ—ó
- –ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Ä—è–¥–æ–∫: "aabcaabxaaaz".
- Z-—Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è —Ü—å–æ–≥–æ —Ä—è–¥–∫–∞ –≤–∏–∑–Ω–∞—á–∞—î –¥–ª—è –∫–æ–∂–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó —ñ–Ω–¥–µ–∫—Å—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –¥–æ–≤–∂–∏–Ω—É –ø—Ä–µ—Ñ—ñ–∫—Å–∞, —è–∫–∏–π –∑–±—ñ–≥–∞—î—Ç—å—Å—è –∑ –ø—ñ–¥—Ä—è–¥–∫–æ–º, —â–æ –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ —Ü—ñ—î—ó –ø–æ–∑–∏—Ü—ñ—ó.

**–ö—Ä–æ–∫–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è Z-—Ñ—É–Ω–∫—Ü—ñ—ó:**
- $Z[0] = 13$, –æ—Å–∫—ñ–ª—å–∫–∏ –≤–µ—Å—å —Ä—è–¥–æ–∫ —î –ø—Ä–µ—Ñ—ñ–∫—Å–æ–º —Å–∞–º–æ–≥–æ —Å–µ–±–µ.
- –î–ª—è $i = 1$ –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –ø—ñ–¥—Ä—è–¥–æ–∫ "abcaabxaaaz". –ù–∞–π–¥–æ–≤—à–∏–π –ø—Ä–µ—Ñ—ñ–∫—Å, —è–∫–∏–π –∑–±—ñ–≥–∞—î—Ç—å—Å—è –∑ –ø–æ—á–∞—Ç–∫–æ–º —Ä—è–¥–∫–∞, –º–∞—î –¥–æ–≤–∂–∏–Ω—É 1 (—Ç–æ–±—Ç–æ —Ü–µ –ø–µ—Ä—à–∞ –ª—ñ—Ç–µ—Ä–∞ "a"). –û—Ç–∂–µ, $Z[1] = 1$
- –î–ª—è $i = 2$ –ø—ñ–¥—Ä—è–¥–æ–∫ "bcaabxaaaz". –ù–∞–π–¥–æ–≤—à–∏–π –ø—Ä–µ—Ñ—ñ–∫—Å –Ω–µ –∑–±—ñ–≥–∞—î—Ç—å—Å—è –∑ –ø–æ—á–∞—Ç–∫–æ–º —Ä—è–¥–∫–∞, —Ç–æ–º—É $Z[2] = 0$.
- –î–ª—è $i = 3$ –ø—ñ–¥—Ä—è–¥–æ–∫ "caabxaaaz". –ê–Ω–∞–ª–æ–≥—ñ—á–Ω–æ, $Z[3] = 0$.
- –î–ª—è $i = 4$ –ø—ñ–¥—Ä—è–¥–æ–∫ "aabxaaaz". –ü—Ä–µ—Ñ—ñ–∫—Å –¥–æ–≤–∂–∏–Ω–æ—é 3 ("aab") –∑–±—ñ–≥–∞—î—Ç—å—Å—è –∑ –ø–æ—á–∞—Ç–∫–æ–º —Ä—è–¥–∫–∞. –û—Ç–∂–µ, $Z[4] = 3$.
- –î–ª—è $i = 5$ –ø—ñ–¥—Ä—è–¥–æ–∫ "abxaaaz". –¢—É—Ç –Ω–µ–º–∞—î –∑–±—ñ–≥—É, —Ç–æ–º—É $Z[5] = 0$.
- –î–ª—è $i = 6$ –ø—ñ–¥—Ä—è–¥–æ–∫ "bxaaaz". –ó–Ω–æ–≤—É –Ω–µ–º–∞—î –∑–±—ñ–≥—É, —Ç–æ–º—É $Z[6] = 0$.
- –î–ª—è $i = 7$ –ø—ñ–¥—Ä—è–¥–æ–∫ "xaaaz". –ü—Ä–µ—Ñ—ñ–∫—Å –¥–æ–≤–∂–∏–Ω–æ—é 2 ("aa") –∑–±—ñ–≥–∞—î—Ç—å—Å—è –∑ –ø–æ—á–∞—Ç–∫–æ–º —Ä—è–¥–∫–∞. –û—Ç–∂–µ, $Z[7] = 2$.
