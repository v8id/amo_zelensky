## <center> Розв'язання </center>

### 1. Вхідні дані

Маємо граф, заданий у вигляді множини вершин і ребер:

- **Вершини**: $V = \{1, 2, 3, 4, 5, 6, 7\}$  
- **Ребра**: $E = \{(1, 2), (1, 3), (2, 4), (3, 5), (4, 5), (4, 6), (6, 7)\}$

Це орієнтований ациклічний граф (DAG). Завдання: знайти топологічний порядок вершин за допомогою алгоритму DFS.

---

### 2. Побудова графа

На основі даних будуємо граф:

- Вершини: додаємо всі елементи з $V$.
- Ребра: додаємо всі пари з $E$.

Граф можна візуалізувати (див. побудоване зображення).

---

### 3. Ідея алгоритму DFS для топологічного сортування

1. Топологічне сортування працює так: вершини графа впорядковуються таким чином, щоб для кожного ребра $(u, v)$ вершина $u$ передувала $v$ у порядку.  
2. Алгоритм обходу в глибину (DFS) дозволяє зручно реалізувати це:
   - Відвідуємо вершину, якщо вона ще не відвідана.
   - Рекурсивно обходимо всіх її сусідів.
   - Після відвідування всіх сусідів додаємо поточну вершину до стеку.
   - У кінці стек розвертається, щоб отримати порядок.

---

### 4. Реалізація алгоритму DFS

#### Кроки:

1. Ініціалізуємо:
   - Порожню множину $visited$ для збереження відвіданих вершин.
   - Порожній список $stack$, де зберігається порядок вершин.
2. Викликаємо DFS для кожної вершини, яка ще не відвідана:
   - Якщо вершина $v$ ще не відвідана:
     - Позначаємо її як відвідану.
     - Для всіх сусідів вершини $v$ викликаємо DFS.
     - Після завершення рекурсії додаємо вершину $v$ до $stack$.
3. У результаті список $stack$ розвертається.

---

### 5. Виконання алгоритму на даному графі

#### Вхідні дані:

- Вершини: $V = \{1, 2, 3, 4, 5, 6, 7\}$  
- Ребра: $E = \{(1, 2), (1, 3), (2, 4), (3, 5), (4, 5), (4, 6), (6, 7)\}$

#### Кроки виконання DFS:

1. Починаємо з вершини $1$:
   - Позначаємо $1$ як відвідану.
   - Йдемо до сусіда $2$:
     - Позначаємо $2$ як відвідану.
     - Йдемо до сусіда $4$:
       - Позначаємо $4$ як відвідану.
       - Йдемо до сусіда $5$: $5$ не має сусідів, додаємо $5$ до $stack$.
       - Йдемо до сусіда $6$:
         - Позначаємо $6$ як відвідану.
         - Йдемо до сусіда $7$: $7$ не має сусідів, додаємо $7$ до $stack$.
       - $6$ завершила обхід, додаємо $6$ до $stack$.
     - $4$ завершила обхід, додаємо $4$ до $stack$.
   - $2$ завершила обхід, додаємо $2$ до $stack$.
   - Йдемо до сусіда $3$:
     - Позначаємо $3$ як відвідану.
     - Йдемо до сусіда $5$: $5$ вже відвідано.
     - $3$ завершила обхід, додаємо $3$ до $stack$.
   - $1$ завершила обхід, додаємо $1$ до $stack$.

2. Обхід завершено. У стеку: $[1, 3, 2, 4, 6, 7, 5]$.

---

### 6. Виведення результату

Розгортаємо стек, щоб отримати топологічний порядок:

\[
[1, 3, 2, 4, 6, 7, 5]
\]

Це і є топологічний порядок графа.


## <center> Контрольні питання </center>

### 1. Які переваги і недоліки алгоритму Кана порівняно з алгоритмом DFS для топологічного сортування графа?
#### Порівняння алгоритмів Кана і DFS для топологічного сортування

#### Алгоритм Кана
##### Переваги:
1. Простий для розуміння та реалізації.
2. Виявляє цикли автоматично (залишаються вершини з ненульовим вхідним степенем).
3. Використовує чергу, що ефективно працює для великих графів.

### Недоліки:
1. Потребує додаткової пам'яті для зберігання вхідних степенів.
2. Менш гнучкий, підходить тільки для DAG.
3. Ітеративний, менш інтуїтивний для аналізу графової структури.

---

#### Алгоритм DFS
##### Переваги:
1. Мінімум пам'яті (немає потреби у вхідних степенях).
2. Гнучкість — можна легко модифікувати для роботи з не-DAG.
3. Рекурсивний підхід краще виявляє графову структуру.

##### Недоліки:
1. Складніше реалізувати через рекурсію.
2. Для виявлення циклів потрібна додаткова логіка.
3. Рекурсія може призводити до переповнення стека для великих графів.

---

####  <center> Характеристики в порівнянні  </center>

| Характеристика            | Алгоритм Кана        | Алгоритм DFS          |
|----------------------------|----------------------|-----------------------|
| **Структура даних**         | Черга               | Стек                  |
| **Виявлення циклів**        | Просте              | Потрібна логіка       |
| **Пам'ять**                 | Додатковий масив    | Мінімум               |
| **Простота реалізації**     | Вища                | Складніша             |
| **Складність**              | $O(V + E)$          | $O(V + E)$            |

---

#### Висновок:
- **Кан**: Простий і підходить для великих DAG.
- **DFS**: Гнучкий, але складніший у реалізації.


### 2. Яка складність часу і пам’яті для кожного з алгоритмів у найгіршому і найкращому випадках?

#### Алгоритм Кана

##### Складність часу:
- **Найгірший випадок**: $O(V + E)$  
  (Перевірка всіх вершин та ребер для обробки вхідних степенів і черги)

- **Найкращий випадок**: $O(V + E)$  
  (Те ж саме, навіть якщо граф вже має деякі відвідані вершини)

##### Складність пам'яті:
- **Найгірший випадок**: $O(V + E)$  
  (Потрібно зберігати вхідні степені та структуру черги)

- **Найкращий випадок**: $O(V + E)$  
  (Потрібно зберігати вхідні степені та структуру черги)

---

#### Алгоритм DFS

##### Складність часу:
- **Найгірший випадок**: $O(V + E)$  
  (Відвідування кожної вершини та ребра)

- **Найкращий випадок**: $O(V + E)$  
  (Те ж саме, навіть якщо граф вже частково оброблений)

##### Складність пам'яті:
- **Найгірший випадок**: $O(V)$  
  (Для зберігання стека рекурсії та масиву відвіданих вершин)

- **Найкращий випадок**: $O(V)$  
  (Мінімальна пам'ять для рекурсивного виклику)



### 3. Чи можна застосовувати алгоритм Кана до графів з вагами на ребрах? Як це порівняти з DFS?

#### Алгоритм Кана:
- **Застосування до графів з вагами на ребрах**:  
  Алгоритм Кана **не використовує ваги на ребрах**. Він працює лише з вхідними степенями та не залежить від ваг. Тому вага ребер не впливає на порядок топологічного сортування.

#### Алгоритм DFS:
- **Застосування до графів з вагами на ребрах**:  
  Алгоритм DFS також **не використовує ваги на ребрах** для сортування вершин, оскільки він базується на відвідуванні вершин у порядку глибини. Ваги можуть бути використані в інших контекстах (наприклад, для пошуку найкоротшого шляху), але не для топологічного сортування.

---

#### Отже:
- **Алгоритм Кана** і **DFS** однаково не використовують ваги на ребрах для топологічного сортування. Обидва алгоритми працюють з орієнтованими ациклічними графами незалежно від ваг.


### 4. Як впливає структура графа на швидкість роботи кожного з цих алгоритмів?

#### Алгоритм Кана:
- **Щільність графа**:
  - У графах з великою кількістю ребер (щільні графи) алгоритм Кана працює ефективно, оскільки потрібно лише обробити вхідні степені та відвідувати кожну вершину і ребро.
  
- **Розріджені графи**:
  - У розріджених графах (де мало ребер) алгоритм Кана також працює добре, оскільки складність залежить лише від кількості вершин та ребер ($O(V + E)$).

#### Алгоритм DFS:
- **Щільність графа**:
  - У щільних графах DFS може бути менш ефективним через більшу кількість рекурсивних викликів, оскільки потрібно обробляти більше ребер, але загальна складність залишатиметься $O(V + E)$.
  
- **Розріджені графи**:
  - У розріджених графах DFS також працює ефективно, з мінімальними витратами на пам'ять та рекурсію.

---

#### Отже:
- **Алгоритм Кана** може бути швидшим для графів з великою кількістю ребер завдяки ітеративному підходу.
- **DFS** може працювати менш ефективно в складних графах через рекурсивний підхід, але його складність залишається лі


### 5. Чи є обмеження використання кожного алгоритму для певних типів графів або завдань?

#### Алгоритм Кана:
- **Обмеження**:
  1. **Тільки для DAG (орієнтовані ациклічні графи)**:
     - Алгоритм Кана призначений лише для графів, які не містять циклів.
  2. **Не працює з графами, де є цикли**:
     - Якщо в графі є цикл, алгоритм не зможе здійснити топологічне сортування.
  3. **Не використовує ваги на ребрах**:
     - Алгоритм Кана не враховує ваги ребер, тому не підходить для задач, де важливо враховувати ваги при сортуванні чи оптимізації.

#### Алгоритм DFS:
- **Обмеження**:
  1. **Тільки для DAG**:
     - Як і алгоритм Кана, DFS підходить лише для DAG, хоча можна модифікувати його для виявлення циклів.
  2. **Не працює з графами з великими глибинами рекурсії**:
     - Для великих графів рекурсія може призвести до переповнення стека.
  3. **Не враховує ваги ребер**:
     - DFS не використовує ваги ребер для сортування чи оптимізації, тому не підходить для задач з ваговими графами.

---

#### Отже:
- **Алгоритм Кана** підходить для задач з топологічним сортуванням в DAG, але не працює з циклами та не враховує ваги.
- **Алгоритм DFS** також підходить для DAG, але може мати обмеження при великих графах через рекурсію та не використовує ваги на ребрах.


### 6. Які варіанти оптимізації можна застосувати для кожного алгоритму з метою поліпшення його продуктивності?

#### Алгоритм Кана:
##### Оптимізації:
1. **Використання структури даних для зберігання вхідних степенів**:
   - Використовувати **хеш-таблиці** для швидкого доступу до вхідних степенів вершин.
   
2. **Черга з пріоритетами**:
   - Можна застосувати **чергу з пріоритетами** (heap), щоб вибирати вершини з нульовим вхідним степенем у порядку їх обробки (але це додатково ускладнює алгоритм).

3. **Паралельна обробка**:
   - Якщо є кілька вершин з нульовим вхідним степенем, можна використовувати **паралельні потоки** для одночасної обробки таких вершин.

#### Алгоритм DFS:
##### Оптимізації:
1. **Ітеративна версія DFS**:
   - Замість рекурсії використовувати **стек** для ітеративного обходу, що дозволяє уникнути переповнення стека при великих графах.

2. **Раннє завершення**:
   - Якщо топологічне сортування не потрібно для всіх вершин, можна **завершити DFS раніше**, як тільки буде знайдено необхідний результат.

3. **Пропуск відвіданих вершин**:
   - Замість повторного відвідування однакових вершин, можна зберігати **масив відвіданих вершин** для пропуску вже оброблених.
