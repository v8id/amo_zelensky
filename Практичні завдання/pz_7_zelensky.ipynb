#### Крок 3: Відновлення підпослідовності

Відновлємо найбільшу спільну підпослідовність, рухаючись від правого нижнього кута таблиці (dp[5][5]) вгору і вліво:

1. dp[5][5] = 1, dp[4][5] = 1 — символи не збігаються.
2. dp[4][4] = 1, dp[3][4] = 1 — символи "L" і "L" збігаються, додаємо "L".
3. dp[3][3] = 1, dp[2][3] = 0 — символи не збігаються.
4. dp[2][2] = 0, dp[1][2] = 0 — символи не збігаються.

Отже, найбільша спільна підпослідовність (LCS) для рядків **"HELLO"** і **"WORLD"** є: **L**

```python
def longest_common_subsequence(s1, s2):
    m = len(s1)
    n = len(s2)
    
    # Створення таблиці для зберігання проміжних результатів
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Заповнення таблиці знизу вгору
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # Виведемо таблицю dp для перевірки
    print("Таблиця dp:")
    for row in dp:
        print(row)
    
    # Відновлення найбільшої спільної підпослідовності
    lcs = []
    i, j = m, n
    while i > 0 and j > 0:
        if s1[i - 1] == s2[j - 1]:
            lcs.append(s1[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] >= dp[i][j - 1]:
            i -= 1
        else:
            j -= 1
    
    # Перевернення lcs, оскільки ми додавали символи з кінця
    lcs.reverse()
    return ''.join(lcs)

# Приклад використання
s1 = "HELLO"
s2 = "WORLD"
result = longest_common_subsequence(s1, s2)
print("Найбільша спільна підпослідовність:", result)
```

## <center> Контрольні питання </center>

### 1. У чому полягає задача знаходження найдовшої спільної підпослідовності (LCS)?

Задача LCS полягає в знаходженні найдовшої підпослідовності, що є спільною для двох заданих послідовностей. Підпослідовність зберігає відносний порядок елементів, але може пропускати деякі елементи.

**Вхід:** Два рядки або послідовності.  
**Вихід:** Найдовша спільна підпослідовність (LCS).

### 2. Які головні методи можна використовувати для знаходження найдовшої спільної підпослідовності?

#### **Алгоритм динамічного програмування (DP):**  
   Використовує таблицю для збереження проміжних результатів і оптимізує розв'язок через розбиття задачі на підзадачі.  
   - Складність: **O(m * n)**, де m і n — довжини двох послідовностей.

#### **Рекурсивний підхід:**  
   Застосовує рекурсію для перевірки всіх можливих підпослідовностей.  
   - Складність: **O(2^n)**, що робить метод неефективним для великих входів.

#### **Алгоритм на основі матриць (матриця LCS):**  
   Створення двовимірної матриці для збереження довжин спільних підпослідовностей.  
   - Використовує аналогічний підхід до алгоритму динамічного програмування, але може бути реалізований із додатковими оптимізаціями для економії пам'яті.

#### **Алгоритм Ратнера (згортання до єдиної послідовності):**  
   Підходить для деяких специфічних випадків задач, де можна згорнути послідовності в єдину.

### 3. Як працює алгоритм динамічного програмування для знаходження LCS?

#### **Ідея:**  
   Алгоритм динамічного програмування розв'язує задачу знаходження LCS, розбиваючи її на менші підзадачі. Для кожної пари індексів (i, j) визначається довжина LCS для підрядків першого та другого рядка до i та j відповідно.

#### **Кроки:**
   - Створюється таблиця dp розміру (m + 1) x (n + 1), де m і n — довжини двох рядків.
   - dp[i][j] зберігає довжину LCS для підрядка s1[0..i-1] і s2[0..j-1].
   - Якщо s1[i-1] == s2[j-1], то dp[i][j] = dp[i-1][j-1] + 1.
   - Якщо s1[i-1] ≠ s2[j-1], то dp[i][j] = max(dp[i-1][j], dp[i][j-1]).
   
#### **Ініціалізація:**  
   dp[0][j] = 0 для всіх j, dp[i][0] = 0 для всіх i (оскільки порожній рядок не має спільної підпослідовності з будь-яким іншим рядком).

#### **Реконструкція LCS:**  
   Після заповнення таблиці dp, відновлюємо LCS, рухаючись з dp[m][n] до dp[0][0] і додаючи символи, коли s1[i-1] == s2[j-1].

#### **Часова складність:**  
   Часова складність алгоритму: **O(m * n)**, де m і n — довжини рядків s1 і s2.

### 4. Як працює алгоритм Хаббарда для знаходження LCS?

Алгоритм Хаббарда — це один з методів, який використовує розширену техніку для знаходження найдовшої спільної підпослідовності (LCS) з меншою кількістю пам'яті порівняно з класичним методом динамічного програмування.

#### **Основна ідея:**  
   Алгоритм Хаббарда оптимізує класичний метод динамічного програмування, зменшуючи використання пам'яті. Він працює в два етапи:
   - Замість збереження всієї таблиці розміром **O(m * n)**, таблиця використовується лише для зберігання поточного та попереднього рядків.
   - Це дозволяє знизити використання пам'яті до **O(min(m, n))**, де m і n — довжини рядків.

#### **Кроки алгоритму:**
   - Використовує дві одночасно оновлювані таблиці: одну для поточного рядка, іншу — для попереднього.
   - Для кожної пари індексів (i, j) обчислюється значення LCS, порівнюючи символи з кожного рядка.
   - Якщо символи збігаються, значення у таблиці оновлюється на основі значення попереднього рядка.
   - Якщо символи не збігаються, вибирається максимальне значення між попереднім рядком і поточним стовпцем.

#### **Переваги:**
   - Зменшує використання пам'яті до **O(min(m, n))** без втрати часу виконання.
   - Придатний для задач з великими рядками, де важлива економія пам'яті.

#### **Недоліки:**
   - Не є дуже зручним для відновлення самої підпослідовності, оскільки зберігається лише інформація про довжину LCS.

#### **Часова складність:**  
   Часова складність алгоритму Хаббарда: **O(m * n)**, де m і n — довжини рядків s1 і s2.


### 5. Які переваги та недоліки алгоритмів динамічного програмування та Хаббарда для знаходження LCS?

#### **Алгоритм динамічного програмування (DP):**

   **Переваги:**
   - **Простота реалізації**: Легко зрозуміти і реалізувати.
   - **Коректність**: Завжди дає правильний результат для всіх можливих входів.
   - **Відновлення LCS**: Легко відновити саму спільну підпослідовність за допомогою таблиці dp.

   **Недоліки:**
   - **Високе використання пам'яті**: Для двох рядків довжиною m і n необхідно зберігати таблицю розміру **O(m * n)**, що може бути проблемою при великих рядках.
   - **Часова складність**: Хоча складність **O(m * n)** є досить ефективною, для дуже великих рядків цей алгоритм може бути повільним.

#### **Алгоритм Хаббарда:**

   **Переваги:**
   - **Економія пам'яті**: Використовує **O(min(m, n))** пам'яті, що є значною перевагою при роботі з великими рядками.
   - **Швидкість**: Часова складність залишається такою ж, як у класичному методі — **O(m * n)**, але зменшення пам'яті дозволяє працювати з більшими наборами даних.

   **Недоліки:**
   - **Складність реалізації**: Порівняно складніше реалізувати, ніж стандартний алгоритм динамічного програмування.
   - **Відновлення LCS**: Відновлення самій LCS може бути ускладненим, оскільки зберігається тільки поточний та попередній рядки, а не вся таблиця.
