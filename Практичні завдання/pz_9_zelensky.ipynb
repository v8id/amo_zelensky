# <center> Практична робота №9 </center>
## **Тема. Стиснення даних. Кодування і декодування Гафмена**
## **Мета:** набути практичних навичок застосування алгоритму оптимального кодування Гафмена.
## **Варіант:** 7

## <center> Завдання </center>

### **7.** Маємо текст AABAABACBBCBCEEFFFGGGLLLRRRTTGGGDEE. Закодувати текст, використовуючи алгоритм Гафмена. Побудувати двійкове дерево. Оцінити ефект від кодування у порівнянні з неоптимальним випадком.

## <center> Розв'язання </center>

#### Неоптимальний випадок:
У неоптимальному випадку символи можуть бути закодовані рівноцінно, наприклад, за допомогою стандартного коду (наприклад, ASCII), де кожен символ займає однакову кількість біт. Це призводить до того, що часто зустрічаються символи займають стільки ж біт, скільки й рідші, що неефективно з точки зору стиснення.

##### Оцінка ефекту:
1. **Кількість біт для зберігання символів**:
   - У випадку стандартного кодування кожен символ займає однакову кількість біт. Наприклад, в ASCII кожен символ займає 8 біт.
   - В алгоритмі Гафмена кожен символ отримує змінну кількість біт, яка пропорційна його частоті. Часті символи використовують менше бітів.

2. **Економія бітів**:
   - Кількість біт, що необхідні для кодування тексту, буде меншою в алгоритмі Гафмена порівняно з стандартним кодуванням, якщо вхідний набір даних має неоднакові частоти символів. Наприклад, якщо в тексті часто зустрічаються кілька символів, то їх можна закодувати короткими кодами.
   
3. **Практичний ефект**:
   - Економія біт може бути значною, залежно від розподілу частот символів. У найкращому випадку, при високій неоднорідності частот, ефективність алгоритму Гафмена може бути на кілька десятків відсотків кращою за неоптимальні методи.

##### Формула для ефективності:
Для оцінки ефективності алгоритму Гафмена можна порівняти середню довжину кодування символу в алгоритмі Гафмена з середньою довжиною символу у стандартному коді.

- **Неоптимальний код**: для кожного символу використовується однакова кількість біт, наприклад, $k$ бітів на символ.
- **Код Гафмена**: середня довжина коду для символу $L_{\text{Huffman}} = \sum_{i} p_i \cdot l_i$, де $p_i$ — ймовірність символу, а $l_i$ — довжина коду символу.

Ефективність алгоритму Гафмена можна оцінити через **коефіцієнт стиснення**:

$$
C = \frac{\text{довжина коду для стандартного коду}}{\text{довжина коду для Гафмена}} = \frac{k}{L_{\text{Huffman}}}
$$

Таким чином, чим більше значення коефіцієнта стиснення, тим більше вигоди від застосування алгоритму Гафмена.

## <center> Контрольні питання </center>

### 1. Що таке кодування Гафмена та як воно працює?

#### Кодування Гафмена — алгоритм стиснення, який мінімізує кількість біт для представлення символів.

##### Алгоритм:
1. **Частотний аналіз:** Підраховуємо частоту кожного символу.
2. **Будуємо дерево:** Найменші частоти об'єднуються в вузли.
3. **Призначення кодів:** Символи з більшою частотою отримують коротші коди.
4. **Стиснення:** Замінюємо символи їх бінарними кодами.

Цей підхід забезпечує ефективне стиснення даних.


### 2. Як визначається оптимальний двійковий код Гафмена для стиснення даних?

#### Оптимальний код Гафмена визначається через:

##### **Частоти символів:** Визначаємо частоту кожного символу в даних.
##### **Будова дерева:** Створюємо бінарне дерево, де:
   - Символи з найменшою частотою об'єднуються в нові вузли.
   - Повторюємо, поки не отримаємо одне дерево.
##### **Призначення кодів:** Кожен символ отримує унікальний код, де:
   - Символи з високою частотою отримують коротші коди.
   - Символи з низькою частотою — довші.
##### **Оптимальність:** Код є оптимальним, оскільки мінімізує середню довжину коду, використовуючи принцип частотного кодування.

Це гарантує мінімальний розмір коду для стиснення.


### 3. Які переваги має кодування Гафмена над іншими методами стиснення даних?

1. **Оптимальність:** Дає мінімальну середню довжину коду для набору даних, що забезпечує найкраще стиснення.
2. **Адаптивність:** Працює навіть з динамічними даними, підлаштовуючи коди під частоту символів.
3. **Простота реалізації:** Алгоритм легко реалізується з використанням стандартних структур даних (наприклад, дерев).
4. **Широке застосування:** Використовується у багатьох форматах стиснення (наприклад, в ZIP, PNG).

Завдяки цим перевагам, кодування Гафмена є одним з найпопулярніших методів для ефективного стиснення даних.


### 4. Як відбувається декодування даних, закодованих за допомогою кодування Гафмена?

#### **Відновлення дерева:** Для декодування необхідно мати дерево Гафмена, яке було побудоване під час кодування.
#### **Перехід по дереву:** Починаємо з кореня дерева і рухаємося:
   - Якщо біт — 0, йдемо в лівий вузол.
   - Якщо біт — 1, йдемо в правий вузол.
#### **Отримання символа:** Коли доходимо до листа дерева, отримуємо символ.
#### **Повторення:** Повторюємо процес для кожного бітового послідовності до повного декодування.

Цей процес гарантує правильне відновлення вихідних даних без втрат.


### 5. Які є можливі недоліки кодування Гафмена?

1. **Потреба в збереженні дерева:** Для декодування необхідно зберігати або передавати дерево Гафмена, що може збільшити обсяг даних.
2. **Неможливість адаптації до змін:** Якщо частоти символів змінюються в процесі обробки даних, дерево потрібно оновлювати або перезбирати.
3. **Простір для малих файлів:** Для малих обсягів даних кодування Гафмена може бути неефективним, оскільки накладні витрати на побудову дерева можуть перевищувати вигоди від стиснення.
4. **Часова складність:** Алгоритм має часову складність $O(n \log n)$, що може бути не оптимальним для великих обсягів даних.

Ці обмеження роблять кодування Гафмена менш ефективним у деяких випадках.


### 6. Для чого використовується побудова дерева в кодуванні Гафмена?

#### Побудова дерева є основою кодування Гафмена, оскільки:

1. **Структурує символи за частотою:** Дерево відображає частоти символів у вигляді бінарного дерева, де найбільш часті символи розташовуються ближче до кореня.
2. **Оптимізує довжину коду:** Дерево допомагає визначити оптимальні бінарні коди для символів, де більш часті символи мають коротші коди, а рідкісні — довші.
3. **Декодування:** Дерево необхідне для точного відновлення вихідних даних при декодуванні — рухаючись по дереву, можна відновити оригінальні символи з бітових послідовностей.

Без дерева неможливо здійснити як кодування, так і декодування в кодуванні Гафмена.
