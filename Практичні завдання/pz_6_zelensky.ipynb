### Алгоритм Беллмана-Форда для пошуку найкоротших шляхів

Маємо граф, в якому потрібно знайти найкоротші шляхи від вершини 1 до всіх інших вершин. Використовуємо алгоритм Беллмана-Форда.

#### Крок 1: Опис графа

Маємо наступні вершини та ребра з вагами:

- $1 \to 2$ (вага 2)
- $1 \to 3$ (вага 5)
- $1 \to 5$ (вага 2)
- $2 \to 4$ (вага 7)
- $3 \to 6$ (вага 5)
- $4 \to 5$ (вага 1)
- $5 \to 2$ (вага 1)
- $5 \to 4$ (вага 1)
- $6 \to 2$ (вага 8)

Будемо знаходити найкоротші шляхи від вершини 1 до всіх інших.

#### Крок 2: Ініціалізація

Нехай відстань до кожної вершини буде ініціалізована як нескінченність, окрім відстані до початкової вершини (1), яку встановимо рівною 0.

- Відстань до вершини 1: $d[1] = 0$
- Відстань до інших вершин: $d[v] = \infty$, де $v$ — це інші вершини.

#### Крок 3: Виконання алгоритму Беллмана-Форда

Алгоритм Беллмана-Форда виконується $V - 1$ ітерацій, де $V$ — кількість вершин у графі (в нашому випадку $V = 6$).

##### Перша ітерація:
- Для кожного ребра перевіряємо, чи можна зменшити відстань до вершини $v$ через вершину $u$.
  - Для ребра $1 \to 2$ (вага 2): $d[2] = \min(\infty, d[1] + 2) = 2$.
  - Для ребра $1 \to 3$ (вага 5): $d[3] = \min(\infty, d[1] + 5) = 5$.
  - Для ребра $1 \to 5$ (вага 2): $d[5] = \min(\infty, d[1] + 2) = 2$.
  - Для ребра $2 \to 4$ (вага 7): $d[4] = \min(\infty, d[2] + 7) = 9$.
  - Для ребра $3 \to 6$ (вага 5): $d[6] = \min(\infty, d[3] + 5) = 10$.
  - Для ребра $4 \to 5$ (вага 1): $d[5] = \min(2, d[4] + 1) = 2$.
  - Для ребра $5 \to 2$ (вага 1): $d[2] = \min(2, d[5] + 1) = 2$.
  - Для ребра $5 \to 4$ (вага 1): $d[4] = \min(9, d[5] + 1) = 3$.
  - Для ребра $6 \to 2$ (вага 8): $d[2] = \min(2, d[6] + 8) = 2$.

Після першої ітерації ми отримуємо наступні відстані:
- $d[1] = 0$
- $d[2] = 2$
- $d[3] = 5$
- $d[4] = 3$
- $d[5] = 2$
- $d[6] = 10$

##### Друга ітерація:
- Повторюємо перевірку для кожного ребра. Оскільки відстані вже оптимізовані, вони не змінюються.
  
##### Третя ітерація:
- Всі відстані залишаються незмінними після третьої ітерації.

І так далі. Алгоритм завершується після $V - 1 = 5$ ітерацій.

#### Крок 4: Перевірка на наявність циклів з від'ємною вагою

Після виконання $V - 1$ ітерацій перевіряємо, чи є ще можливість зменшити відстань. Якщо це можливо, то в графі є цикл з від'ємною вагою. У нашому випадку циклів з від'ємною вагою немає.

#### Крок 5: Результати

Найкоротші відстані від вершини 1 до інших вершин:
- Відстань від 1 до 1: 0
- Відстань від 1 до 2: 2
- Відстань від 1 до 3: 5
- Відстань від 1 до 4: 3
- Відстань від 1 до 5: 2
- Відстань від 1 до 6: 10

## <center> Контрольні питання </center>

### 1. Що таке граф і які головні складові його структури?

Граф — це математична структура, що складається з набору вершин (також відомих як вузли) та набору ребер (або зв'язків) між ними.

#### Основні складові графа:

##### 1. **Вершини (V)**: 
   - Вершини є основними об'єктами графа, які можуть представляти будь-які сутності (наприклад, точки, об'єкти або організації).
   
##### 2. **Ребра (E)**:
   - Ребра є з'єднаннями між вершинами, що можуть бути орієнтованими або неорієнтованими.
   - **Орієнтовані графи**: Ребра мають напрямок від однієї вершини до іншої.
   - **Неорієнтовані графи**: Ребра не мають напрямку, вони з'єднують вершини в обох напрямках.
   
##### 3. **Вага (optional)**: 
   - У деяких графах ребра можуть мати вагу (наприклад, відстань або вартість), що додає додаткову інформацію про зв'язки між вершинами.
   
##### 4. **Структура графа**:
   - **Ациклічні графи**: Графи без циклів.
   - **Циклічні графи**: Графи, що містять цикли.

### 3. Як працює алгоритм Дейкстри і які його особливості?

#### Принцип роботи:
1. **Ініціалізація**:
   - Встановлюється початкова відстань до початкової вершини рівною 0, а до всіх інших вершин — нескінченність.
   - Мітки всіх вершин позначаються як неперевірені.

2. **Вибір вершини з найменшою відстанню**:
   - З множини неперевірених вершин вибирається вершина з найменшою поточною відстанню.
   
3. **Оновлення відстаней**:
   - Для всіх сусідніх вершин поточної вибраної, якщо через поточну вершину можна знайти коротший шлях, оновлюється відстань.

4. **Повторення кроків**:
   - Повторюється кроки 2 і 3, поки всі вершини не будуть перевірені.

5. **Завершення**:
   - Алгоритм завершується, коли всі вершини оброблені або коли мінімальна відстань до незвіданої вершини нескінченна (що означає відсутність шляху).

#### Особливості:
1. **Використання черги з пріоритетами**:
   - Для вибору вершини з найменшою відстанню алгоритм часто використовує **чергу з пріоритетами** (heap), що дозволяє знизити складність вибору мінімального елементу до $O(\log V)$.

2. **Обмеження на ваги ребер**:
   - Алгоритм Дейкстри працює тільки з **невід'ємними вагами** на ребрах. Якщо є від'ємні ваги, алгоритм може працювати неправильно.

3. **Одноразовий запуск**:
   - Алгоритм знаходить найкоротший шлях від однієї вершини до всіх інших, а не від усіх до всіх (це робить алгоритм Флойда-Уоршелла).

#### Складність часу:
- **За допомогою черги з пріоритетами**: $O((V + E) \log V)$.


### 4. Що таке алгоритм Белмена–Форда і коли його варто застосовувати?

#### Алгоритм Белмена-Форда: принцип роботи та застосування

#### Принцип роботи:
1. **Ініціалізація**:
   - Встановлюється відстань до початкової вершини рівною 0, а до всіх інших вершин — нескінченність.
   
2. **Ітерації**:
   - Алгоритм виконує $V-1$ ітерацій, де в кожній ітерації перевіряються всі ребра графа.
   - Для кожного ребра $(u, v)$, якщо відстань до вершини $v$ через $u$ менша, ніж поточна відстань, то відстань оновлюється.

3. **Перевірка на від'ємні цикли**:
   - Після $V-1$ ітерацій алгоритм виконує ще одну перевірку. Якщо є ребра, де можна ще оновити відстань, то граф містить від'ємний цикл.

#### Коли застосовувати:
1. **Графи з від'ємними вагами**:
   - Алгоритм Белмена-Форда є єдиним популярним методом для знаходження найкоротших шляхів у графах з **від'ємними вагами** на ребрах.

2. **Виявлення від'ємних циклів**:
   - Алгоритм також використовується для виявлення **від'ємних циклів** в графах, які можуть бути важливими для задач оптимізації або теорії графів.

3. **Менша ефективність у порівнянні з Дейкстрою**:
   - Алгоритм Белмена-Форда менш ефективний за **алгоритм Дейкстри** для графів з невід'ємними вагами, оскільки його складність більша ($O(VE)$), але він підходить для випадків, коли є від'ємні ваги.

#### Складність часу:
- **Часова складність**: $O(VE)$, де $V$ — кількість вершин, а $E$ — кількість ребер.

### 5. Як працює алгоритм Флойда–Форшала і які його переваги та недоліки?

#### Алгоритм Флойда-Форшала: принцип роботи та переваги/недоліки

#### Принцип роботи:
1. **Ініціалізація**:
   - Створюється матриця відстаней між усіма парами вершин. Спочатку відстань між кожною парою вершин $i$ та $j$ дорівнює ваги ребра між ними, або нескінченність, якщо ребра немає. Відстань між кожною вершиною і самою собою дорівнює нулю.

2. **Ітерації**:
   - Алгоритм виконує трійку вкладених циклів:
     - Для кожної вершини $k$ перевіряються всі пари вершин $i$ і $j$.
     - Якщо відстань через вершину $k$ є коротшою, ніж поточна відстань між $i$ та $j$, то відстань оновлюється:
       $$
       d(i, j) = \min(d(i, j), d(i, k) + d(k, j))
       $$

3. **Завершення**:
   - Після виконання всіх ітерацій матриця відстаней містить найкоротші шляхи між усіма парами вершин.

#### Переваги:
1. **Всі пари найкоротших шляхів**:
   - Алгоритм знаходить найкоротші шляхи між усіма парами вершин, що робить його універсальним для задач, де потрібна така інформація.

2. **Підтримка від'ємних ваг**:
   - Алгоритм Флойда-Форшала працює з графами, які містять від'ємні ваги на ребрах, якщо тільки в графі немає від'ємних циклів.

3. **Простота**:
   - Алгоритм є простим у реалізації, оскільки використовує лише матрицю відстаней та три вкладених цикли.

#### Недоліки:
1. **Часова складність**:
   - Часова складність алгоритму становить $O(V^3)$, що робить його неефективним для дуже великих графів.
   
2. **Пам'ятевитратність**:
   - Алгоритм вимагає зберігання матриці $V \times V$, що може бути непрактичним для великих графів.

3. **Відсутність оптимізації для одиночних шляхів**:
   - Якщо потрібно знайти лише найкоротший шлях між однією парою вершин, то алгоритм Флойда-Форшала буде неефективним у порівнянні з іншими методами, такими як алгоритм Дейкстри.

#### Складність часу:
- **Часова складність**: $O(V^3)$, де $V$ — кількість вершин у графі.
